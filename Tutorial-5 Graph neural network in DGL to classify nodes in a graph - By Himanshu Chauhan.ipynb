# Message Passing

*** Message Passing ***

- One common operation for updating node features is message passing, where the source nodes send messages through edges to the destinations. With DGLGraph, we can do this with send() and recv()

-  The nodes send information computed via the message functions, and aggregates incoming information
with the reduce functions.



** In this node classification example : each node will update its own feature with information sent from neighboring nodes** 


- At layer l, each node vli carries a feature vector hli.
- Each layer of the GCN tries to aggregate the features from uli where ui‘s are neighborhood nodes to v into the next layer representation at vl+1i. This is followed by an affine transformation with some non-linearity.

```python
# Define the message & reduce function
# NOTE: we ignore the GCN's normalization constant c_ij for this tutorial.
def gcn_message(edges):
    # The argument is a batch of edges.
    # This computes a (batch of) message called 'msg' using the source node's feature 'h'.
    return {'msg' : edges.src['h']}

#Just like you need to go to your mailbox for retrieving mails, 
#the destination nodes also need to receive the messages and potentially update their features

def gcn_reduce(nodes):
    # The argument is a batch of nodes.
    # This computes the new 'h' features by summing received 'msg' in each node's mailbox.
    # and replacing the original feature.
    return {'h' : torch.sum(nodes.mailbox['msg'], dim=1)}
```

# Step 3: Defining a Graph Convolutional Network  ( GCN )

```python
import torch.nn as nn
import torch.nn.functional as F
```

*** Defining the GCNLayer module***

CLASS torch.nn.Module
Base class for all neural network modules.

Your models should also subclass this class.

Modules can also contain other Modules, allowing to nest them in a tree structure. You can assign the submodules as regular attributes:



```python
# Use the nn package to define our model as a sequence of layers.
# nn.Sequential is a Module which contains other Modules, and applies them in sequence to produce its output.

class GCNLayer(nn.Module):
    
    def __init__(self, in_features, out_features): # Initialize our neural network before we add more layers
        super(GCNLayer, self).__init__() #super(type, obj): obj must be an instance or subtype of type
        self.linear = nn.Sequential(nn.Linear(in_features, out_features))

    def forward(self, g, inputs):
        # g is the graph and the inputs is the input node features
        # first set the node features
        g.ndata['h'] = inputs
        
        # trigger message passing on all edges
        g.send(g.edges(), gcn_message)
        
        # trigger aggregation at all nodes
        g.recv(g.nodes(), gcn_reduce)
        
        # get the result node features
        h = g.ndata.pop('h')
        
        # perform linear transformation after getting the result node features
        return self.linear(h)
```

*** Define a deeper GCN model that contains two GCN layer ***

```python
# Define a 2-layer GCN model
class GCN(nn.Module):
    def __init__(self, in_features, hidden_size, num_classes_output):
        super(GCN, self).__init__() #super(type, obj): obj must be an instance or subtype of type
        self.gcn1 = GCNLayer(in_features, hidden_size)
        self.gcn2 = GCNLayer(hidden_size, num_classes_output)

    def forward(self, g, inputs):
        h = self.gcn1(g, inputs)
        h = torch.relu(h)
        h = self.gcn2(g, h)
        return h
```

```python
# The first layer transforms input features of size of 34 to a hidden size of 5.
# The second layer transforms the hidden layer and produces output features of
# size 2, corresponding to the two groups of the karate club.

in_size = 34
hidden_size = 5
out_size =2
```

# Step 5: Train the Model

The training loop is exactly the same as other PyTorch models. 

(1) create an optimizer, 

(2) feed the inputs to the model, 

(3) calculate the loss and 

(4) use autograd to optimize the model.

```python
import torch.optim as optim
```

```python
net4 = GCN(in_size,hidden_size,out_size) # Instantiating the model by callingthe GCN Class
optimizer4 = torch.optim.Adam(net4.parameters(), lr=0.01)
```

```python
all_logits = []
for epoch in range(40):
    Y_pred = net4(g, input_features)
    
    # we save the logits for visualization later
    all_logits.append(Y_pred.detach())
    log_predict = F.log_softmax(Y_pred, 1)
    
    _, pred = torch.max(Y_pred.data, -1)
    
    # we only compute loss for labeled nodes
    loss = F.nll_loss(log_predict[labeled_nodes], labels)

    optimizer4.zero_grad() # zero out the gradients for fresh calculations
    loss.backward() #call loss.backward() to calulate gradients
    optimizer4.step()  # update the parameters by applying gradients
 
    print('Epoch %d | Loss: %.4f ' % (epoch, loss.item()) )
    print("Predict lables for all the Nodes in the graph:\n")
    print(pred)
```
